## Q2. 請使用 python 畫中山大學管理學院周邊地圖！

提示：

2. 使用 folium 套件來顯示地圖。

```python
!pip install geocoder
!pip install folium
```

```python
import folium;import geocoder
```

```python
loc=geocoder.google('中山大學管理學院') #只會給出地址
print(loc)
loc=geocoder.google('中山大學管理學院').latlng #給出經緯
print(loc)
folium.Map(loc,zoom_start=17) #zoom_start越大，可以較小範圍看地圖
```
